```python
import numpy as np
import scipy.special as ss
import matplotlib.pyplot as plt
```

```python
L = 15

numx = 1025

xvec = np.linspace(-L, L, numx)

ychebvec = ss.eval_chebyt(10, xvec/L)

plt.plot(xvec, ychebvec, '.-')
plt.show()
```

```python
yexpvec1 = np.exp(-(xvec+4)**2 / 20)
yexpvec2 = -np.exp(-(xvec-5)**2 / 20)

plt.plot(xvec, yexpvec1, '.-')
plt.plot(xvec, yexpvec2, '.-')
plt.show()

ycmplxvec = yexpvec1 + 1j*yexpvec2

plt.plot(xvec, np.abs(ycmplxvec), '.-')
plt.show()
```
